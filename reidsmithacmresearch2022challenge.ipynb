```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt # data visualization, graphs
import sklearn as sk

# Input data files are available in the read-only "../input/" directory

data = pd.read_csv('/kaggle/input/carsforsale/cars_raw.csv')

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session

#Convert String of Price to Int
def data_price_to_int(dt_val) -> int:
    if dt_val[0] == "N":
        return None
    return int(dt_val[1:].replace(",",""))

#Replace Price Column with Int Values
data['Price'] = list(map(data_price_to_int, data['Price']))
```

```python
#Car Color v. Price Chart 

#Make Color -> Price_List_of_Color Dictionary
bar_dict = {}
for ind, car in data.iterrows():
    if car['ExteriorColor'] in bar_dict:
        bar_dict[car['ExteriorColor']].append(car['Price'])
    else:
        bar_dict[car['ExteriorColor']] = [car['Price']]

#Sort Data by Color Names Alphabetically
color_names = sorted(list(bar_dict.keys()))
color_prices = [bar_dict[color] for color in color_names]


#Color v. Price Box Plot
box_plot = plt.boxplot(color_prices, labels=color_names, showmeans=True, showfliers=False)
plt.title("Prices of Cars by Color")
plt.xlabel("Car Exterior Color")
plt.ylabel("Car Price ($)")

#Show Color v. Price Box Plot
plt.show()
print("Displayed Color v. Price Box Plot")

#Sort Color Names by Mean (Simple Version)

#Python One 1-liner go BRRRRRR
simple_mean_list = sorted([(color_names[i],box_plot['means'][i].get_ydata()[0]) for i in range(len(color_names))], key = lambda x: x[1])

#Human Readable Version of Above
"""
simple_mean_list = []
for i in range(len(color_names)):
    simple_mean_list.append((color_names[i],box_plot['means'][i].get_ydata()[0]))
simple_mean_list.sort(key=lambda x: x[1])
"""

#Results of Simple Mean Sorting
print("Simple Mean Sorting:")
print("5 Cheapest Car Colors on Average: ",end="")
print(*simple_mean_list[0:5],sep=", ")
print("\n5 Most Expensive Car Colors on Average: ",end="")
print(*simple_mean_list[-5:-1],sep=", ")

#Sort Color Names by Mean (Slightly More Reasonable Version)
#Slightly More Reasonable = The Color Has Been Sold More Than 4 Times

#Python One 1-liner go BRRRRRR
better_mean_list = sorted([(color_names[i],box_plot['means'][i].get_ydata()[0]) for i in range(len(color_names)) if len(bar_dict[color_names[i]]) > 4], key=lambda x: x[1])
#Human Readable Version of Above
"""
better_mean_list = []
for i in range(len(color_names)):
    if len(bar_dict[color_names[i]]) > 4:
        better_mean_list.append((color_names[i],box_plot['means'][i].get_ydata()[0]))
better_mean_list.sort(key=lambda x: x[1])
"""

#Results of Slightly More Reasonable Mean Sorting
print("\n\nBetter Mean Sorting:")
print("5 Cheapest Car Colors on Average: ",end="")
print(*better_mean_list[0:5],sep=", ")
print("\n5 Most Expensive Car Colors on Average: ",end="")
print(*better_mean_list[-5:-1],sep=", ")


```
